# Data is cleaned and ready for Regression analysis

```python
#Spliting the data 
x = data2.drop('TempHighF' , axis = 1).values
y = data2['TempHighF'].values
```

```python
X_train , X_test , Y_train , Y_test = train_test_split(x,y, test_size = 0.3 , random_state = 0)
```

```python
# Print the first 5 elements of X_train
print(X_train[:5])  # This selects elements from index 0 (inclusive) to 4 (exclusive)
```

```python
lnr = LinearRegression()
lnr.fit(X_train, Y_train)
```

```python
y_pred = lnr.predict(X_test)
```

```python
import pickle
```

```python
# File paths for model and DataFrame
model_filepath = r"C:\Users\jay\Downloads\Group project\Webpage\savedmodels\actual_vs_predicted_df.pickle"
df_filepath = r"C:\Users\jay\Downloads\Group project\Webpage\savedmodels\linear_regression_model.pickle"

# Save the model
with open(model_filepath, 'wb') as f:
    pickle.dump(lnr, f)

# Save the DataFrame
with open(df_filepath, 'wb') as f:
    pickle.dump(data2, f)
```

```python
df = pd.DataFrame({'Actual': Y_test, 'Predicted': y_pred})

df.head()
```

```python
import seaborn as sns
sns.pairplot(df)
```

# From the plot our model looks like a very good fit for this data 

```python
# Checking the accuracy  of the model 

print("Accuracy of the model =",lnr.score(X_train, Y_train))
print("Mean absolute error =", round(sm.mean_absolute_error(Y_test, y_pred),2)) 
print("Mean squared error =", round(sm.mean_squared_error(Y_test, y_pred), 2)) 
print("Median absolute error =", round(sm.median_absolute_error(Y_test, y_pred), 2)) 
print("Explain variance score =", round(sm.explained_variance_score(Y_test, y_pred), 2)) 
print("R2 score =", round(sm.r2_score(Y_test, y_pred), 2))
```
